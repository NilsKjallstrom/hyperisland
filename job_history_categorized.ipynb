```python
import pandas as pd
```

```python
#Dropping columns which are sensitive and exessive
df =df.drop(columns= 'Dokumentnr')
df =df.drop(columns= 'Namn')
```

```python
#Dropping invoice and related credit invoice which has been repaid(2 rows)
df = df.drop(index= [5,6])
```

```python
#fixing index after removing 2 rows
df = df.reset_index(drop=True)
```

```python
#fetching unique "benämingar" to eventually categorize the services. 

unique_benamningar = df['Benämning'].unique().tolist()
print(unique_benamningar)
```

```python
# creating function to categorize services based on essence of "benämningar"

def classify_article(article_name):
    if 'isättning' in article_name:
        return 'Indoor painting'
    elif 'Fönster' in article_name:
        return 'Outdoor painting'
    elif article_name == 'Dörrmålning':
        return 'Outdoor painting'
    elif article_name == 'måla om klädkammare':
        return 'Indoor painting'
    elif article_name == 'Bredspackling':
        return 'Indoor painting'
    elif article_name == 'Grundarbete-Rivning':
        return 'Indoor painting'
    elif article_name == 'Grundarbete-Slipning-Snickerier':
        return 'Indoor painting'
    elif article_name == 'Grundarbete-Rivning':
        return 'Indoor painting'
    elif article_name == 'Grundarbete-Rivning-Bortforsling':
        return 'Indoor painting'
    elif article_name == 'Skrapning':
        return 'Indoor painting'
    elif article_name == 'Lås innedörrar':
        return 'Indoor painting'
    elif article_name == 'trappmålning':
        return 'Indoor painting'
    elif article_name == 'slipning-spackling av dörrar':
        return 'Indoor painting'
    elif article_name == 'spackling av skarvar och ovanför dörr':
        return 'Indoor painting'
    elif article_name == 'Målning runt fönster/dörrar (fasad)':
        return 'Outdoor painting'
    elif article_name == 'Färdigmålning staket + fasad på sida och baksida av huset':
        return 'Outdoor painting'
    elif article_name == 'Grundmålning staket + fasad på sida och baksida av hus':
        return 'Outdoor painting'
    elif 'spackling' in article_name:
        return 'Indoor painting'
    elif article_name == 'Takmålning':
        return 'Indoor painting'
    elif article_name == 'Väggmålning':
        return 'Indoor painting'
    elif 'Grundmålning' in article_name:
        return 'Indoor painting'
    elif 'Fasad' in article_name:
        return 'Outdoor painting'
    elif 'Färdigmålning' in article_name:
        return 'Indoor painting'
    elif 'Fönster' in article_name:
        return 'Outdoor painting'
    elif 'gipsning' in article_name:
        return 'Indoor painting'
    elif 'luckor' in article_name:
        return 'Indoor painting'
    elif 'Målning' in article_name:
        return 'Indoor painting'
    elif 'Spackling' in article_name:
        return 'Indoor painting'
    elif 'Material' in article_name:
        return 'Materials'
    elif 'Patentering' in article_name:
        return 'Indoor painting'
    elif 'Tapetsering' in article_name:
        return 'Indoor painting'
    elif 'Inne' in article_name:
        return 'Indoor painting'
    else:
        return 'Other' 
    
    
    
```

```python
#Changing the dtype to str, enable function above to work
df['Benämning'] = df['Benämning'].astype(str)


```

```python
#applying function
df['Category'] = df['Benämning'].apply(classify_article)
```
